```python
import matplotlib.pyplot as plt
def color_plots(mix_one, mix_two, mixture):
    colors = ["black"]*25
    c = data["odorNames"].index(mix_one)
    h = data["odorNames"].index(mix_two)
    ch = data["odorNames"].index(mixture)
    colors[c] = "red"
    colors[h] = "blue"
    colors[ch] = "purple"
    plt.scatter(*coords.T, c = colors)
```
